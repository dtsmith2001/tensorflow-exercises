````python
a = tf.Variable(tf.random_uniform([10,10]), name='a')
````

````python
b = tf.Variable(tf.random_uniform([10,10]), name='b')
````

````python
c = tf.matmul(a, b, name='ab')
````

````python
sess = tf.Session(config=tf.ConfigProto(log_device_placement=True))
````

````python
writer = tf.train.SummaryWriter('/tmp/', sess.graph)
````

````python
sess.run(tf.initialize_all_variables())
````

````python
tf.reset_default_graph()
````

````python
sess.run(c)
````

````python
import numpy as np
````

````python
a = np.random.rand(10,10)
b = np.random.rand(10,10)
x = tf.placeholder(tf.float32, [10, 10])
y = tf.placeholder(tf.float32, [10, 10])
c = tf.matmul(a, b, name='ab')
sess = tf.Session(config=tf.ConfigProto(log_device_placement=True))
writer = tf.train.SummaryWriter('/tmp/', sess.graph)
sess.run(tf.initialize_all_variables())
````

## sess.run

````python
sess.run(c, feed_dict={x:a,y:b})
````

````python
tf.reset_default_graph()
````

````python
W = tf.Variable(tf.random_uniform([1000,1000]))
````

````python
x = tf.Variable(tf.ones([1000,1]))
````

````python
sess = tf.Session()
writer = tf.train.SummaryWriter('/tmp/', sess.graph)
````

````python
sess.run(tf.matmul(W, x))
````

````python
sess.run(tf.initialize_all_variables())
````

````python
sess.run(tf.matmul(W, x))
````
